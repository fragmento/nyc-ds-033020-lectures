```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris.head(3)
```

Lets use seaborn to visualize the data.

```python
%matplotlib inline
sns.set()
sns.pairplot(iris, hue='species', height=1);
```

Feature Matrix and target array

```python
X_iris = iris.drop('species', axis=1)
X_iris.shape
```

```python
Y_iris = iris['species']
Y_iris.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.naive_bayes import GaussianNB
```

to instantiate the linearRegression class and specify that we would like to fit  the intercept using the fit_intercept hyperparameter

```python
model_lin = LinearRegression(n_jobs=-1)
model_lin
```

Spliting the data into traininig set and testing set

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X_iris, Y_iris, random_state=1)
```

```python
model_GB = GaussianNB()
```

```python
model_GB.fit(Xtrain,ytrain)
```

```python
y_model_GB = model_GB.predict(Xtest)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_model_GB, ytest)
```
